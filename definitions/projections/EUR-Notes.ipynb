# Notes on the EUR domain

Considering the data given here: https://cordex.org/domains/

We would expect to be able to calculate the lower left and upper right corners of the domain from regular latlong to rotated latlong. However this is proving difficult because it is not immediately clear if we are considering the center of the lower left pixel (or upper right pixel), or the corners.

For that reason, the code below is kept for further inquiries if ever needed. For pyku purposes, we use the data provided directly with the ``py-cordex`` library

### References

* https://cordex.org/domains/

* https://euro-cordex.net/060378/index.php.en

* https://is-enes-data.github.io/cordex_archive_specifications.pdf

* https://github.com/WCRP-CORDEX/domain-tables/blob/main/CORDEX-CMIP6_domain_boundaries.csv

```python
import pyproj
from pyresample import utils

# Define lower left and upper right corners in geographic coordinates
# -------------------------------------------------------------------

lon_ll = +350.01
lat_ll = +22.2
lon_ur = +64.4
lat_ur = +66.65

# Define projection string and transformer
# ----------------------------------------

proj_string = (
    "+proj=ob_tran +o_proj=longlat +o_lat_p=39.25 +o_lon_p=0 +lon_0=18"
)

transformer = pyproj.Transformer.from_crs(
    'EPSG:4326',
    proj_string,
    always_xy=True
)

# Determine lower left and upper right corners in projection coordinates
# ----------------------------------------------------------------------

x_ll, y_ll = transformer.transform(lon_ll, lat_ll)
x_ur, y_ur = transformer.transform(lon_ur, lat_ur)

x_ll, y_ll = round(x_ll, 4), round(y_ll, 4)
x_ur, y_ur = round(x_ur, 4), round(y_ur, 4)
```
