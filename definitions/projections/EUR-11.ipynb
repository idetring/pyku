## Rotated longlat projection definition

```python
import cordex as cx
import pyproj
from pyresample import utils
```

```python
proj_string = (
    "+proj=ob_tran +o_proj=longlat +o_lat_p=39.25 +o_lon_p=0 +lon_0=18"
)
```

```python
import cordex as cx

domain_info = cx.domain_info('EUR-11')
domain_info
```

```python
# Set number of pixels
# --------------------

x_size = domain_info['nlon']
y_size = domain_info['nlat']

x_ll, y_ll = domain_info['ll_lon']-domain_info['dlon']/2., domain_info['ll_lat']-domain_info['dlat']/2.
x_ur, y_ur = domain_info['ur_lon']+domain_info['dlon']/2., domain_info['ur_lat']+domain_info['dlat']/2.

# Create pyresample.AreaDefinition
# --------------------------------

area_id = domain_info['short_name']
area_name = (
    "EUR-11 rotated longitude latitude grid in accordance with CORDEX domains "
    "requirements https://cordex.org/domains/"
)

proj_id = None
proj4_args = proj_string

area_extent = (x_ll, y_ll, x_ur, y_ur)
area_def = utils.get_area_def(
    area_id,
    area_name,
    proj_id,
    proj4_args,
    x_size,
    y_size,
    area_extent
)

print(area_def.dump())
```
